## 1 Side scheme

### 1.1 Case 1: 3-dimensional harmonic oscillator

* potential: $V(r) = m \omega^2 r^2 / 2$
* simulation parameters:
  * run time: $10^5\ \text{fs}$
  * step size: $2.418884326509 \times 10^{-3}\ \text{fs}$ ($0.1\ \text{a.u.}$)
  * data collection period: 1000
* system information: $m = 1\ \text{a.u.}$, $\omega = 1\ \text{a.u.}$, $\beta = 8\ \text{a.u.}$

The position distribution of the first degree of freedom:

```python
import numpy as np
import matplotlib.pyplot as plt

chkfile = "./test_at_side_ho_e/test_at_side_ho_e_01.chk"

times, positions, momenta = np.loadtxt(chkfile, unpack=True)

n, bins, patches = plt.hist(positions, bins='auto', density=True, facecolor='lightskyblue', label='Num. Simu.')
plt.title("Position distribution")
plt.xlabel("Position (a.u.)")
plt.ylabel("Probability density")

x = np.linspace(min(positions), max(positions), num=1000)
y = np.exp(-4 * np.power(x, 2))/np.sqrt(0.25 * np.pi)

plt.plot(x, y, linestyle='solid', c='dodgerblue', lw=3, alpha=0.8, label='Anal. Curv.')
plt.legend(loc='best', frameon=False)
plt.show()
```

The momentum distribution of the first degree of freedom:

```python
import numpy as np
import matplotlib.pyplot as plt

chkfile = "./test_at_side_ho_e/test_at_side_ho_e_01.chk"

times, positions, momenta = np.loadtxt(chkfile, unpack=True)

n, bins, patches = plt.hist(momenta, bins='auto', density=True, facecolor='lightskyblue', label='Num. Simu.')
plt.title("Momentum distribution")
plt.xlabel("Momentum (a.u.)")
plt.ylabel("Probability density")

x = np.linspace(min(momenta), max(momenta), num=1000)
y = np.exp(-4 * np.power(x, 2))/np.sqrt(0.25 * np.pi)

plt.plot(x, y, linestyle='solid', c='dodgerblue', lw=3, alpha=0.8, label='Anal. Curv.')
plt.legend(loc='best', frameon=False)
plt.show()
```

The evolution of kinetic energy:

```python
import numpy as np
import matplotlib.pyplot as plt

outfile = "./test_at_side_ho_e/test_at_side_ho_e_01.out"
times, kin_enes, pot_enes = np.loadtxt(outfile, unpack=True)

plt.plot(times, kin_enes, c='lightskyblue')
plt.title("Kinetic energy versus time")
plt.xlabel("Time (fs)")
plt.ylabel("Energy (a.u.)")
plt.show()
```

The evolution of potential energy:

```python
import numpy as np
import matplotlib.pyplot as plt

outfile = "./test_at_side_ho_e/test_at_side_ho_e_01.out"
times, kin_enes, pot_enes = np.loadtxt(outfile, unpack=True)

plt.plot(times, pot_enes, c='lightskyblue')
plt.title("Potential energy versus time")
plt.xlabel("Time (fs)")
plt.ylabel("Energy (a.u.)")
plt.show()
```
